# Comparison Analysis of Sentiment Analysis Models

## Objective
The objective of this analysis is to compare the performance of three deep learning models
for sentiment analysis on the IMDb Movie Reviews dataset:
1. Custom LSTM trained from scratch
2. Pretrained AWD-LSTM fine-tuned using ULMFiT
3. Transformer-based BERT model

The comparison focuses on accuracy, convergence speed, and computational cost.


## Model Performance Comparison

| Model | Epochs | Accuracy | Observation |
|------|-------|---------|-------------|
| Custom LSTM | 5 | ~0.50 | Slow convergence, weak baseline |
| AWD-LSTM (ULMFiT) | 1 | 0.839 | Fast convergence using transfer learning |
| BERT | 2 | ~0.90+ | Best contextual understanding |


## Analysis and Interpretation

The custom LSTM model trained from scratch struggled to learn complex linguistic
patterns from long movie reviews, resulting in poor accuracy even after multiple epochs.
This highlights the difficulty of training deep NLP models without pretrained knowledge.

The AWD-LSTM model, fine-tuned using the ULMFiT methodology, achieved significantly higher
accuracy after just a single epoch. This demonstrates the effectiveness of transfer learning
in capturing syntactic and semantic patterns.

The BERT model achieved the highest accuracy among all models. Its bidirectional
self-attention mechanism enables superior contextual understanding, allowing it to
outperform recurrent architectures on long-text sentiment classification tasks.


## Conclusion

This comparison demonstrates a clear progression in NLP model performance.
While custom LSTM models provide foundational understanding, pretrained models
such as AWD-LSTM and BERT significantly improve both accuracy and convergence speed.

BERT achieved the best overall performance, while AWD-LSTM offered an effective
balance between accuracy and computational efficiency. These results highlight
the importance of transfer learning and transformer architectures in modern NLP systems.
